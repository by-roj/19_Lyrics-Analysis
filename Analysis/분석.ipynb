```python
erase_list = ["1940's 이전 베스트 가요 콜렉션","1950's 베스트 가요 콜렉션","1960's 베스트 가요 콜렉션",
             "1960~1963년 베스트 가요 콜렉션","1970's 베스트 가요 콜렉션","1980's 베스트 가요 콜렉션",
             "1990's 베스트 가요 콜렉션","2000's 베스트 가요 콜렉션"]

bugs = bugs[~bugs['year'].isin(erase_list)]
bugs['year'] = bugs['year'].apply(lambda x: x[:4])
```

```python
revised_time_list = []

for item in bugs['runtime']:
    splitted = item.split(':')
    minute = int(splitted[0][1])*60
    seconds = int(splitted[1])
    revised_time = minute+seconds
    revised_time_list.append(revised_time)
    
bugs['runtime'] = revised_time_list
```

```python
groupma = ["올스타 밴드","에밀레","야생마","홍삼트리오","템페스트","휘버스 (열기들)","샌드 페블즈(Sand Pebbless)"]
groupfe = ["쿨 시스터즈","이 시스터즈","정 시스터즈","이씨스터즈","화니 시스터즈(Fanny Sisters)"]
grouphon = ["이건주와 어린이들","하눌타리","E.G.B","퓨처라이거"]
duoma = ["하사와 병장"]
duofe = ["작품 하나"]
soloma = ["하성관","이예성","유춘산","도성아","이신","정연택","신세영","한정무","허민"]
solofe = ["나현희","정난이","허림","조정희","정은숙","이정민","심연옥"]

for i in range(len(bugs)) :
    a = bugs[i:i+1]
    if a.singer.item() in groupma :
        bugs.iloc[i]["type"] = "그룹 (남성)"
    if a.singer.item() in groupfe :
        bugs.iloc[i]["type"] = "그룹 (여성)"
    if a.singer.item() in grouphon :
        bugs.iloc[i]["type"] = "그룹 (혼성)"
    if a.singer.item() in duoma :
        bugs.iloc[i]["type"] = "듀오 (남성)"
    if a.singer.item() in duofe :
        bugs.iloc[i]["type"] = "듀오 (여성)"
    if a.singer.item() in soloma :
        bugs.iloc[i]["type"] = "솔로 (남성)"
    if a.singer.item() in solofe :
        bugs.iloc[i]["type"] = "솔로 (여성)"
```

```python
bugs["sex"] = bugs.type.str[-3:-1]
bugs["type"] = bugs.type.str[0:2]

bugs["lyrics"] = bugs.lyric.str.replace('\r\n',' ')
del bugs["lyric"]
```

```python
# 일단 연도별로 합치기
years = bugs.year.unique()
```

```python
bugs_full = []
for i in range(0,len(years)) :
    x = bugs[bugs.year == years[i]]
    length = len(x)
    xx = x.lyrics[x.lyrics.index[0]]
    for j in range(1,length) :
        xx = xx + x.lyrics[x.lyrics.index[j]]
    bugs_full.append(xx)
```

```python
len(bugs_full)
```

```python
from konlpy.tag import Okt; t = Okt() # open korean text
pos = lambda d: ['/'.join(p) for p in t.pos(d, stem=True)]
texts_ko = [pos(doc) for doc in bugs_full]
```

```python
import gensim
from gensim import models
from gensim import corpora
dictionary_ko = corpora.Dictionary(texts_ko)
```

```python
tf_ko = [dictionary_ko.doc2bow(text) for text in texts_ko]
tfidf_model_ko = models.TfidfModel(tf_ko)
tfidf_ko = tfidf_model_ko[tf_ko]
```

```python
def tfidf_to_words(index, n_words):
    id_list = [sorted(tfidf_ko[index], key=lambda x: x[1], reverse=True)[:n_words][i][0] for i in range(n_words)]
    word_list = [dictionary_ko[id_list[i]] for i in range(n_words)]
    return word_list 
```

```python
for i in range(0,len(years)) :
    print(tfidf_to_words(len(years)-i-1, 10))
```

## 사랑에 대한 인식의 변화

```python
# 각 가사별로 명사만 뽑게 만들기
nouns = [t.nouns(doc) for doc in bugs_full]
```

```python
stop_words = ['너','내','나','난','날','그','그대','널','수','더','니','것','네','이','넌','걸','또','다시','왜','때','이제','젠','못','안']
```

```python
import nltk

for i in range(0,len(years)) :
    allnouns = [each_word for each_word in nouns[len(years)-i-1] if each_word not in stop_words]
    ko = nltk.Text(allnouns,name='blah blah')
    ko.similar('사랑',3)
```
