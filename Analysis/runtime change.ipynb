```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
bugs = pd.read_csv("./Data_Preprocessing/Bugs_preprocessing.csv")
```

```python
bugs.head()
```

```python
runtime_sum = [0] * 55
count = [0] * 55
```

```python
for i in range(len(bugs)):
    runtime_sum[2018 - bugs.loc[i, 'year']] += bugs.loc[i, 'runtime']
    count[2018 - bugs.loc[i, 'year']] += 1
```

```python
runtime_avg = pd.DataFrame(columns=['year', 'runtime_avg'])
```

```python
for i in range(len(runtime_sum)):
    runtime_avg.loc[i, 'year'] = 2018 - i
```

```python
for i in range(len(runtime_sum)):
    runtime_avg.loc[i, 'runtime_avg'] = runtime_sum[i] / count[i]
```

```python
runtime_avg.tail(30)
```

```python
plt.plot(runtime_avg['year'], runtime_avg['runtime_avg'], color='blue', linestyle='solid')
plt.title('Runtime change')
plt.xlabel('year')
plt.ylabel('runtime average')
plt.show()
```
